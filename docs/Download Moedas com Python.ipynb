# VectorBT - Crypto

### Para instalação, execute:

```python
#!pip install python-binance
#!pip install vectorbt
```

### Exemplo de código:

```python
import vectorbt as vbt

# Definir o símbolo BTCUSDT
symbol = 'BTCUSDT'


# Baixar dados com intervalos de 15 minutos usando VectorBT
cripto_data = vbt.BinanceData.download(
    symbol,
    start='7 day ago UTC', 
    end='now UTC',
    interval='15m'
)

# Exibir os dados
data = cripto_data.get()
data.tail()
```
